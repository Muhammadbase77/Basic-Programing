## 🔹 ตัวอย่างที่ 1: ใช้ NumPy คำนวณข้อมูล NDVI จากภาพ Raster


```python

import numpy as np

#สร้าง “ภาพจำลอง” ขนาด 3×3 pixel (Normalized Difference Vegetation Index)
nir = np.array([[0.7, 0.8, 0.6], [0.9, 0.5, 0.3], [0.4, 0.7, 0.2]])
red = np.array([[0.3, 0.4, 0.2], [0.5, 0.3, 0.1], [0.2, 0.3, 0.1]])

# คำนวณค่า NDVI  ความเขียวของพืช
ndvi = (nir - red) / (nir + red)
#NIR (Near Infrared) → พืชสะท้อนสูง Red → พืชดูดกลืน

print("ค่า NDVI:")
print(ndvi)
```

1. **แบบฝึกหัด 1**: ใช้ NumPy คำนวณค่า **Mean, Max, Min** ของค่า NDVI ในอาร์เรย์ที่สร้างขึ้นเอง

```python
import numpy as np

# Create a "simulated" 3×3 pixel image (Normalized Difference Vegetation Index)
nir = np.array([[0.7, 0.8, 0.6],
                [0.9, 0.5, 0.3],
                [0.4, 0.7, 0.2]])

red = np.array([[0.3, 0.4, 0.2],
                [0.5, 0.3, 0.1],
                [0.2, 0.3, 0.1]])

# Calculate NDVI (plant greenness)
ndvi = (nir - red) / (nir + red)
# NIR (Near Infrared) → plants reflect strongly
# Red → plants absorb strongly

# Show NDVI
print("NDVI values:")
print(ndvi)

# Calculate statistics
mean_ndvi = np.mean(ndvi)
max_ndvi = np.max(ndvi)
min_ndvi = np.min(ndvi)

print(f"Mean NDVI: {mean_ndvi}")
print(f"Max NDVI: {max_ndvi}")
print(f"Min NDVI: {min_ndvi}")
```

3. **แบบฝึกหัด 3**: ใช้ GeoPandas ทำ **Spatial Join** ระหว่างข้อมูลจังหวัดและอำเภอ

```python
import geopandas as gpd

# 1. Load province boundaries (ADM1)
gdf_provinces = gpd.read_file(
    "https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip"
)

# 2. Load district boundaries (ADM2) – you need a link or local path for ADM2!
gdf_districts = gpd.read_file(
    "/content/drive/MyDrive/Code Moo/geoBoundaries-THA-ADM2_simplified.geojson"
)

# 3. Make sure both GeoDataFrames use the same CRS
gdf_districts = gdf_districts.to_crs(gdf_provinces.crs)

# 4. Perform the spatial join
#    Using predicate="within" so each district gets its province info
gdf_joined = gpd.sjoin(
    gdf_districts,
    gdf_provinces,
    how="inner",
    predicate="within"   # within means districts inside province polygons
)

# 5. Show first few rows
print(gdf_joined.head())
```

```python
import geopandas as gpd
import folium
from folium.features import GeoJsonTooltip
import random

# Load provinces (ADM1)
gdf_provinces = gpd.read_file(
    "https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip"
)

# Load districts (ADM2)
gdf_districts = gpd.read_file(
    "/content/drive/MyDrive/Code Moo/geoBoundaries-THA-ADM2_simplified.geojson"
)

# Ensure CRS match
gdf_districts = gdf_districts.to_crs(gdf_provinces.crs)

# Spatial join
gdf_joined = gpd.sjoin(
    gdf_districts,
    gdf_provinces,
    how='inner',
    predicate='within'
)

# Convert to WGS84
gdf_joined = gdf_joined.to_crs(epsg=4326)

# Keep only necessary columns
gdf_clean = gdf_joined[['shapeName', 'ADM1_EN', 'geometry']].reset_index(drop=True)

# Assign random colors to each province
provinces = gdf_clean['ADM1_EN'].unique()
colors = {}
for prov in provinces:
    # random pastel color
    r = random.randint(100, 255)
    g = random.randint(100, 255)
    b = random.randint(100, 255)
    colors[prov] = f'#{r:02x}{g:02x}{b:02x}'

# Create Folium map
m = folium.Map(location=[13.5, 101], zoom_start=6, tiles='CartoDB positron')

# Add districts colored by province
def style_function(feature):
    return {
        'fillColor': colors[feature['properties']['ADM1_EN']],
        'color': 'black',
        'weight': 0.5,
        'fillOpacity': 0.6
    }

folium.GeoJson(
    gdf_clean,
    style_function=style_function,
    tooltip=GeoJsonTooltip(
        fields=['shapeName', 'ADM1_EN'],
        aliases=['District:', 'Province:'],
        localize=True
    )
).add_to(m)

# Convert 'date', 'validOn', 'validTo' columns in gdf_provinces to string to avoid JSON serialization error
for col in ['date', 'validOn', 'validTo']:
    if col in gdf_provinces.columns:
        gdf_provinces[col] = gdf_provinces[col].astype(str)

# Overlay province borders (thicker)
folium.GeoJson(
    gdf_provinces.to_crs(epsg=4326),
    style_function=lambda x: {'color': 'black', 'weight': 2, 'fillOpacity': 0}
).add_to(m)

# Display map
m
```

4. **แบบฝึกหัด 4**: ใช้ NumPy และ GeoPandas ร่วมกันเพื่อหาข้อมูลจังหวัดที่มี NDVI เฉลี่ยสูงสุด

**Notice - These numbers do not come from satellites or real measurements.**
As you can see by this " gdf_joined['NDVI'] = np.random.rand(len(gdf_joined)) ". It’s just that my NDVI values here are simulated for practice.

```python
import geopandas as gpd
import numpy as np

# Load provinces (ADM1)
gdf_provinces = gpd.read_file(
    "https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip"
)

# Load districts (ADM2)
gdf_districts = gpd.read_file(
    "/content/drive/MyDrive/Code Moo/geoBoundaries-THA-ADM2_simplified.geojson"
)

# Ensure CRS match
gdf_districts = gdf_districts.to_crs(gdf_provinces.crs)

# Spatial join: assign each district to a province
gdf_joined = gpd.sjoin(
    gdf_districts,
    gdf_provinces,
    how='inner',
    predicate='within'
)

# Keep only useful columns
gdf_joined = gdf_joined[['shapeName', 'ADM1_EN', 'geometry']].reset_index(drop=True)

# Screates random numbers between 0 and 1.
np.random.seed(42)  # reproducible
gdf_joined['NDVI'] = np.random.rand(len(gdf_joined))

# Show sample
print(gdf_joined[['shapeName', 'ADM1_EN', 'NDVI']].head())

# Group by province
province_ndvi = gdf_joined.groupby('ADM1_EN')['NDVI'].mean().reset_index()

# Sort descending to find highest NDVI
province_ndvi = province_ndvi.sort_values(by='NDVI', ascending=False)

print("Provinces sorted by average NDVI:")
print(province_ndvi.head(5))

max_ndvi_province = province_ndvi.iloc[0]
print(f"Province with highest average NDVI: {max_ndvi_province['ADM1_EN']}")
print(f"Average NDVI: {max_ndvi_province['NDVI']:.3f}")
```

## More example to make it clear.
 ### ***Notice*** :
 I used simulated NDVI data to demonstrate the workflow and test the calculations, allowing us to practice raster-vector analysis without needing large satellite files.

```python
import geopandas as gpd
import rasterio
from rasterio.transform import from_origin
import numpy as np
import pandas as pd
from rasterio.mask import mask

# 1. Load provinces
gdf_provinces = gpd.read_file(
    "https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip"
)
gdf_provinces = gdf_provinces.to_crs(epsg=4326)

# 2. Simulate a raster covering Thailand
width, height = 100, 80  # small raster for demo
xmin, ymin, xmax, ymax = 97, 5, 107, 20  # bounding box for Thailand
pixel_size_x = (xmax - xmin) / width
pixel_size_y = (ymax - ymin) / height
transform = from_origin(xmin, ymax, pixel_size_x, pixel_size_y)

# Create random NDVI values between 0.1 and 0.9
ndvi_data = np.random.uniform(0.1, 0.9, size=(height, width)).astype('float32')

# Save simulated raster to temporary file
raster_path = "/content/ndvi_simulated.tif"
with rasterio.open(
    raster_path, 'w',
    driver='GTiff',
    height=height,
    width=width,
    count=1,
    dtype='float32',
    crs='EPSG:4326',
    transform=transform
) as dst:
    dst.write(ndvi_data, 1)

# 3. Open the raster
raster = rasterio.open(raster_path)

# 4. Calculate average NDVI per province
province_ndvi_list = []

for idx, province in gdf_provinces.iterrows():
    try:
        out_image, out_transform = mask(raster, [province.geometry], crop=True)
        out_image = out_image.astype('float32')

        # Ignore nodata
        mask_valid = ~np.isnan(out_image)
        mean_ndvi = out_image[mask_valid].mean()

        province_ndvi_list.append({
            'province': province['ADM1_EN'],
            'mean_ndvi': mean_ndvi
        })
    except Exception as e:
        print(f"Error for {province['ADM1_EN']}: {e}")

# 5. Convert to DataFrame and sort
df_ndvi = pd.DataFrame(province_ndvi_list)
df_ndvi = df_ndvi.sort_values(by='mean_ndvi', ascending=False)

# 6. Show top provinces
print("Top 5 provinces by average NDVI (simulated):")
print(df_ndvi.head())

# 7. Highest NDVI province
top_province = df_ndvi.iloc[0]
print(f"\nProvince with highest NDVI: {top_province['province']}")
print(f"Average NDVI: {top_province['mean_ndvi']:.3f}")
```
